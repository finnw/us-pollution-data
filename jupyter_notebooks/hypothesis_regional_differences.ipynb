# Regional Differences in US Pollution

**Hypothesis:** There are significant differences in average pollutant levels between regions of the United States (West, East, Other).

**Method:**
- Group states into 'West', 'East', and 'Other' using full state names.
- For each region, calculate the mean value of each pollutant.
- Visualize the average pollutant levels by region using a bar chart.
- This approach allows us to compare pollution across regions and test the hypothesis.

```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_parquet('../data/cleaned_enriched.parquet')
# Define regions using full state names
west = ['California', 'Colorado', 'Washington', 'Oregon', 'Nevada', 'Arizona', 'New Mexico', 'Utah', 'Idaho', 'Montana', 'Wyoming']
east = ['New York', 'Pennsylvania', 'New Jersey', 'Massachusetts', 'Connecticut', 'Maryland', 'Virginia', 'North Carolina', 'South Carolina', 'Georgia', 'Florida']
df['region'] = df['State'].apply(lambda x: 'West' if x in west else ('East' if x in east else 'Other'))
pollutants = [c for c in df.columns if c.endswith(' AQI') or c.endswith(' Mean')]
means = df.groupby('region')[pollutants].mean().T
means.plot(kind='bar')
plt.title('Average Pollutant Levels by Region')
plt.ylabel('Mean Value')
plt.show()
```

## Conclusion
- The bar chart shows average pollutant levels for each region (West, East, Other), grouped by full state names.
- Differences in bar heights indicate which regions have higher or lower mean values for each pollutant.
- If one region consistently has higher bars, it supports the hypothesis of regional differences in pollution.
- If regions are missing, it may be due to lack of data for those states.
- Further analysis could include statistical tests or controlling for confounding factors such as population or industrial activity.
